```python
# Train Logistic Regression model for Churn Prediction

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc

# ✅ If your target ('Churn') is already numeric (0 and 1), no need to map again
# combine_df['Churn'] = combine_df['Churn'].map({'Yes': 1, 'No': 0})  # ❌ remove this line

# Drop non-informative columns
X = combine_df.drop(columns=['Churn', 'customerID'])
y = combine_df['Churn']

# ---- Handle datetime columns ----
datetime_cols = X.select_dtypes(include=['datetime64[ns]']).columns
for col in datetime_cols:
    X[col] = X[col].astype('int64') // 10**9   # convert datetime to numeric (seconds since epoch)

# ---- Identify column types ----
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

# ---- Preprocessor ----
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'  # keep numeric columns as is
)

# ---- Transform features ----
X_encoded = preprocessor.fit_transform(X)

# ---- Split data ----
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)

# ---- Train Logistic Regression model ----
log_reg = LogisticRegression(max_iter=1000, solver='lbfgs')
log_reg.fit(X_train, y_train)

# ---- Predictions ----
y_pred = log_reg.predict(X_test)
y_pred_proba = log_reg.predict_proba(X_test)[:, 1]

# ---- Evaluate model performance ----
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))  # ✅ fixed
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))

# ---- ROC and AUC ----
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)
print("AUC:", roc_auc)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc
from xgboost import XGBClassifier
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier

# Assuming X_encoded (features) and y (target) are prepared as before

# Split dataset (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)

# Train Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predict and evaluate Random Forest
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)[:, 1]

# Train XGBoost model
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)

# Predict and evaluate XGBoost
y_pred_xgb = xgb_model.predict(X_test)
y_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]

# Train LightGBM model
lgb_model = lgb.LGBMClassifier(random_state=42)
lgb_model.fit(X_train, y_train)

# Predict and evaluate LightGBM
y_pred_lgb = lgb_model.predict(X_test)
y_proba_lgb = lgb_model.predict_proba(X_test)[:, 1]

# Metric printing function
def print_metrics(name, y_true, y_pred, y_proba):
    print(f"--- {name} Model ---")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    print("ROC AUC:", auc(fpr, tpr))
    print()

# Compare all three models
print_metrics("Random Forest", y_test, y_pred_rf, y_proba_rf)
print_metrics("XGBoost", y_test, y_pred_xgb, y_proba_xgb)
print_metrics("LightGBM", y_test, y_pred_lgb, y_proba_lgb)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder


# Clean numeric columns
combine_df['TotalCharges'] = pd.to_numeric(combine_df['TotalCharges'], errors='coerce')
combine_df.fillna(combine_df.mean(numeric_only=True), inplace=True)

# Encode categorical columns
cat_cols = combine_df.select_dtypes(include=['object']).columns
for col in cat_cols:
    combine_df[col] = LabelEncoder().fit_transform(combine_df[col])

# Scale features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(combine_df[['MonthlyCharges','TotalCharges','tenure']])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Determine optimal number of clusters
inertia = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 10), inertia, marker='o')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.show()

# Apply clustering
kmeans = KMeans(n_clusters=4, random_state=42)
combine_df['Segment'] = kmeans.fit_predict(scaled_features)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.6, min_samples=5)
anomaly_labels = dbscan.fit_predict(scaled_features)
combine_df['Anomaly'] = anomaly_labels
```

```python
segment_summary = combine_df.groupby('Segment')[['MonthlyCharges','TotalCharges','tenure']].mean()
print(segment_summary)

combine_df.to_csv("clustered_customers.csv", index=False)
```
